```python
import pandas as pd
```

```python
df = pd.read_csv('d/primaryteachers.csv')
#look for comparison percentage of femail teachers
p = df[df['Units of measurement']=='Percent']
p = p.drop_duplicates(subset="Reference Area", keep='first', inplace=False)
#p2014 = p[p["Time Period"]==2014]
pd.options.display.max_rows = 999
#turning it around, percentage of male primary teachers
p['Percent male teachers'] = round((p['Observation Value'] - 100) * -1,1)
del p['Sex']
del p['Age group']
del p['Units of measurement']
del p['Observation Value']
p = p.sort_values('Percent male teachers', ascending=False).reset_index()
del p['index']
```

```python
p.info()
```

```python
#Getting iso codes
iso = pd.read_csv('d/isocodes.csv', header=1)
iso = iso[['Country name[9]',"Alpha-3 code[9]"]].copy()
```

```python
iso.info()
```

```python
p = p.merge(iso, how='outer', left_on='Reference Area', right_on='Country name[9]')
```

```python
p.to_csv('d/percent_maleteachers.csv', index=False)
```
